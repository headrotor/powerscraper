```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime as dt

# Load the CSV file into a Pandas DataFrame

df = pd.read_csv('normalizedPOUS.csv')
#df = pd.read_csv('normalized_smol.csv')

# Convert Unix epoch column to datetime objects

#Convert datetime to Unix epoch
df['datetime2']=pd.to_datetime(df['datetime'], format='%Y-%m-%d %H:%M:%S')
#df['epoch'] = (df['datetime'] - dt.datetime(1970,1,1)).dt.total_seconds()


for col in df.columns:
    print(col)
    
counties = df['CountyName']
counties = list(set(counties))
counties = sorted(counties)
    
print(counties)
    # make and save smol.csv for testing

```

```python

# resample each county and make a new dataframe of percentage outages

resampled_outages = {}
resampled_percent = {}

for county in sorted(counties):
#for county in ['Alameda']:

    print(f"resampling {county}")
    county_data = df.loc[df['CountyName'] == county]
    cd = county_data.drop_duplicates(subset='datetime2', keep='first', inplace=False, ignore_index=True)
    #cust = (df['CustomersOut'] / df['TotalCustomers']) * 100

    percent = cd['CustomersOut']/cd['CustomersTracked']
    percent.index =  pd.to_datetime(cd['datetime2']) 
    resampled_percent[county] = percent.resample('T').ffill()

    outages = cd['CustomersOut']
    outages.index =  pd.to_datetime(cd['datetime2']) 
    resampled_outages[county] = outages.resample('T').ffill()

p_df = pd.DataFrame(resampled_percent)
p_df = p_df.fillna(0)

o_df = pd.DataFrame(resampled_outages)
o_df = o_df.fillna(0)



```

```python
# make new dataframe with zero-one signals if county is above threshold

df2 = pd.DataFrame()
for county in counties:
#for county in ['Alameda']:
    df2[county] = p_df[county] > 0.10
    df2[county] = df2[county].astype(int) 

    startlocs = np.where(np.diff(df2[county]) == 1)
    stoplocs = np.where(np.diff(df2[county]) == -1)
    print(county, len(stoplocs[0]), len(stoplocs[0]))
    #print(df2.iloc[startlocs].index)
```

```python
#make outages CSV

# Open the text file for writing
file_path = "outages.csv"
with open(file_path, mode='w') as text_file:
    # Write the header line
    header = "outage number,Start time,End time,counties\n"
    text_file.write(header)

    # Loop through the data and write each row
    for i, iloc in enumerate(allstart):
        test = df2.iloc[allstart[i] + 1] == 1

        line =  f"{i:3d}, "
        line += f"{str(dfmax.iloc[allstart[i]].name)[:-3]}, {str(dfmax.iloc[allstop[i]].name)[:-3]},"
        line += f" {' '.join(test[test == True].index.values)}\n"
        text_file.write(line)

print(f"Data has been written to {file_path}")
```
